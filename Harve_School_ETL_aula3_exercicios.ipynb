# 📅 PARTE 1 — TRANSFORMANDO DATA

**Conceito:** Timestamp representa uma data em números de segundos a partir de `1970-01-01T00:00Z` — o chamado "instante zero".

Operações com datas seguem a lógica:
- `DATETIME = DATETIME + TIMEDELTA`  → ex: `14/02/2023 = 12/02/2023 + (2 dias)`
- `TIMEDELTA = DATETIME - DATETIME`  → ex: `(2 dias) = 14/02/2023 - 12/02/2023`

Funções úteis:
- `datetime.now()` → data e hora atual
- `timedelta(days=N)` → representa um intervalo de N dias
- `strftime('%s')` → converte datetime para segundos (timestamp)
- `datetime.utcfromtimestamp(N)` → converte timestamp inteiro para datetime

### 🏋️ EXERCÍCIO 1 — Data atual em timestamp
**Objetivo:** Recupere a data e hora atual e exiba no formato datetime.

````python
# 🏋️ EXERCÍCIO 1
# Importe datetime e timedelta do módulo datetime
# Recupere a data atual e armazene em uma variável chamada 'data_atual'
# Imprima o resultado

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 1
from datetime import datetime, timedelta

data_atual = datetime.now()
print(f'Data e hora atual: {data_atual}')
````

### 🏋️ EXERCÍCIO 2 — Subtraindo dias de uma data
**Objetivo:** Se você pegar a data atual e diminuir 90 dias dela, qual será a data resultado?

````python
# 🏋️ EXERCÍCIO 2
# Use datetime.now() e timedelta para subtrair 90 dias da data atual
# Armazene o resultado em 'data_90_dias_atras'
# Imprima o resultado formatado

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 2
from datetime import datetime, timedelta

data_90_dias_atras = datetime.now() - timedelta(90)
print(f'Data atual:          {datetime.now().strftime("%d/%m/%Y")}')
print(f'90 dias atrás:       {data_90_dias_atras.strftime("%d/%m/%Y")}')
````

### 🏋️ EXERCÍCIO 3 — Convertendo datetime para timestamp em milissegundos
**Objetivo:** Converta a data de 5 dias atrás para timestamp em milissegundos (formato usado por APIs como Binance).

💡 Dica: use `strftime('%s')` para segundos e adicione `'000'` para milissegundos.

````python
# 🏋️ EXERCÍCIO 3
# Calcule a data de 5 dias atrás
# Converta para timestamp em milissegundos (inteiro)
# Imprima o resultado

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 3
from datetime import datetime, timedelta

data_5_dias = datetime.now() - timedelta(5)
timestamp_ms = int(data_5_dias.strftime('%s')) * 1000

print(f'Data (5 dias atrás):         {data_5_dias.strftime("%d/%m/%Y %H:%M:%S")}')
print(f'Timestamp em milissegundos:  {timestamp_ms}')
````

````python
# 🏋️ EXERCÍCIO 4
# Use datetime.utcfromtimestamp() para converter o timestamp abaixo
# Imprima a data e hora legível

timestamp_segundos = 1702387987

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 4
import datetime

timestamp_segundos = 1702387987
data_convertida = datetime.datetime.utcfromtimestamp(timestamp_segundos)

print(f'Timestamp: {timestamp_segundos}')
print(f'Data/hora: {data_convertida}')
````

# 🏷️ PARTE 2 — IDENTIFICANDO COLUNAS

**Conceito:** Ao carregar dados de uma API ou JSON, muitas vezes o DataFrame não vem com nomes de colunas. Precisamos renomeá-las usando `df.columns = [...]`.

Também usamos `df.iloc[:, :N]` para selecionar apenas as N primeiras colunas.

### Setup da Parte 2 — Carregar dados do clima via API HG Brasil

````python
# Setup: buscar previsão do tempo de São Paulo
import requests
import pandas as pd

url_clima = 'https://api.hgbrasil.com/weather?key=3b9b39d6&city_name=Sao%20Paulo'
resposta_clima = requests.get(url_clima)
dados_clima = resposta_clima.json()

# Extrair a lista de previsões
previsoes = dados_clima['results']['forecast']
df_clima = pd.DataFrame(previsoes)

print('Colunas originais:', list(df_clima.columns))
print(f'\nTotal de registros: {len(df_clima)}')
df_clima.head(3)
````

### 🏋️ EXERCÍCIO 5 — Renomear colunas do DataFrame de previsão do tempo
**Objetivo:** Traduza as colunas do dataframe da previsão do tempo para português.

Colunas originais (em ordem): `date`, `weekday`, `max`, `min`, `cloudiness`, `rain`, `rain_probability`, `wind_speedy`, `description`, `condition`

💡 Use `df.columns = [lista de nomes]`

````python
# 🏋️ EXERCÍCIO 5
# Crie uma cópia do df_clima chamada df_clima_pt
# Renomeie as colunas para português
# Exiba o resultado com .head()

# Sugestão de nomes:
# date          → 'data'
# weekday       → 'dia_semana'
# max           → 'temp_max'
# min           → 'temp_min'
# cloudiness    → 'nebulosidade'
# rain          → 'chuva'
# rain_probability → 'prob_chuva'
# wind_speedy   → 'vel_vento'
# description   → 'descricao'
# condition     → 'condicao'

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 5
df_clima_pt = df_clima.copy()

df_clima_pt.columns = [
    'data',
    'dia_semana',
    'temp_max',
    'temp_min',
    'nebulosidade',
    'chuva',
    'prob_chuva',
    'vel_vento',
    'descricao',
    'condicao'
]

print('✅ Colunas renomeadas com sucesso!')
df_clima_pt.head()
````

# 🔧 PARTE 3 — TRATANDO CONTEÚDOS DATAFRAME

**Conceito:** Após renomear colunas, muitas vezes os tipos de dado não estão corretos (`object` ao invés de `float` ou `int`). Precisamos:
- Converter tipos com `.astype(tipo)`
- Arredondar valores com `round(df['coluna'].astype(float), casas_decimais)`

### 🏋️ EXERCÍCIO 6 — Arredondar a coluna chuva
**Objetivo:** Arredonde a coluna `chuva` do `df_clima_pt` retirando a casa decimal (0 casas decimais).

💡 Use `round(df['coluna'].astype(float), 0)`

````python
# 🏋️ EXERCÍCIO 6
# Verifique o tipo atual da coluna 'chuva' com .dtypes
# Converta para float e arredonde para 0 casas decimais
# Atualize a coluna no df_clima_pt
# Imprima o resultado

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 6
print('Tipo antes:', df_clima_pt['chuva'].dtype)
print('Valores antes:\n', df_clima_pt['chuva'].head())

df_clima_pt['chuva'] = round(df_clima_pt['chuva'].astype(float), 0)

print('\nTipo depois:', df_clima_pt['chuva'].dtype)
print('Valores depois:\n', df_clima_pt['chuva'].head())
````

### 🏋️ EXERCÍCIO 7 — Descobrir o dia com maior velocidade do vento
**Objetivo:** Crie um código que descobre qual é o dia da semana em que a velocidade do vento estará mais alta.

💡 Dicas:
- Converta `vel_vento` para float (pode conter texto como `'15 km/h'`, então use `str.extract` se necessário)
- Use `.idxmax()` para encontrar o índice do maior valor
- Acesse o nome do dia com `df.loc[idx, 'dia_semana']`

````python
# 🏋️ EXERCÍCIO 7
# Descubra qual é o dia com a maior velocidade do vento
# Imprima: o nome do dia, a data e a velocidade

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 7
# A coluna vel_vento pode vir como string tipo '15 km/h'
# Extraímos só o número com str.extract
df_clima_pt['vel_vento_num'] = df_clima_pt['vel_vento'].str.extract(r'(\d+\.?\d*)').astype(float)

idx_max_vento = df_clima_pt['vel_vento_num'].idxmax()
dia_max_vento  = df_clima_pt.loc[idx_max_vento, 'dia_semana']
data_max_vento = df_clima_pt.loc[idx_max_vento, 'data']
vel_max        = df_clima_pt.loc[idx_max_vento, 'vel_vento']

print(f'🌬️  Dia com maior vento: {dia_max_vento}')
print(f'📅  Data: {data_max_vento}')
print(f'💨  Velocidade: {vel_max}')
````

# 🏆 PARTE 4 — BOOTCAMP (Desafio Final)

**Contexto:** Usando a API financeira do HG Brasil, retorne a conversão de hoje de real para dólar.
Permita ao usuário digitar o valor em reais via `input()` e imprima o valor convertido em dólar.

**API:** `https://api.hgbrasil.com/finance?key=3b9b39d6`

**Navegação no JSON:**
```
resposta['results']['currencies']['USD']['buy']  → taxa de compra do dólar
```

### 🏋️ EXERCÍCIO 8 — Conversor Real → Dólar com input do usuário

````python
# 🏋️ EXERCÍCIO 8
# 1. Faça a requisição na API HG Brasil
# 2. Extraia a taxa de compra do dólar (USD buy)
# 3. Peça ao usuário um valor em reais com input()
# 4. Calcule e imprima o valor equivalente em dólar

# Seu código aqui:
````

````python
# ✅ RESOLUÇÃO 8
import requests

# 1. Requisição à API
url = 'https://api.hgbrasil.com/finance?key=3b9b39d6'
resposta = requests.get(url).json()

# 2. Extrair taxa do dólar
taxa_dolar = resposta['results']['currencies']['USD']['buy']
print(f'💵 Taxa de compra do dólar hoje: R$ {taxa_dolar}')

# 3. Input do usuário
valor_reais = float(input('\nDigite o valor em reais que deseja converter: R$ '))

# 4. Cálculo e resultado
# Para converter BRL → USD dividimos pelo valor do dólar
valor_dolares = valor_reais / taxa_dolar

print(f'\n✅ R$ {valor_reais:.2f} equivale a US$ {valor_dolares:.2f}')
````
